# Time warping de cada neuronio

```python
import pickle
from spikeHelper.loadSpike import loadSpikeBehav, normEpoch

rat = 7
data = pickle.load(open('Data/r'+str(rat)+'_bin100_sigma100.pickle','rb'))

filename = 'Data/rato'+str(rat)+'tudo.mat'
spikes, behavior = loadSpikeBehav(filename)

normed = normEpoch(spikes,behavior,rat,sigma=100)
```

```python
from spikeHelper.dataOrganization import XyTfromEpoch
cubicNorm = XyTfromEpoch(normed)
cubicCrop = XyTfromEpoch(data,getBins=[5,15])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
norm= cubicNorm[:,:,:50]
crop= cubicCrop[:,:,:50]
from sklearn.decomposition import PCA
dimensions ={'norm':[], 'crop':[]}

for i in range(cubicNorm.shape[0]):
    if norm[i,:,:].sum()>0 and crop[i,:,:].sum()>0:
        pcaN = PCA()
        pcaN.fit(norm[i,:,:])
        dimensions['norm'].append([np.cumsum(pcaN.explained_variance_ratio_)])

        pcaC = PCA()
        pcaC.fit(crop[i,:,:])
        dimensions['crop'].append([np.cumsum(pcaC.explained_variance_ratio_)])
```

```python
norm= cubicNorm[:,:,100:150]
crop= cubicCrop[:,:,100:150]
from sklearn.decomposition import PCA
dimensionsLate ={'norm':[], 'crop':[]}

for i in range(cubicNorm.shape[0]):
    if norm[i,:,:].sum()>0 and crop[i,:,:].sum()>0:
        pcaN = PCA()
        pcaN.fit(norm[i,:,:])
        dimensionsLate['norm'].append([np.cumsum(pcaN.explained_variance_ratio_)])

        pcaC = PCA()
        pcaC.fit(crop[i,:,:])
        dimensionsLate['crop'].append([np.cumsum(pcaC.explained_variance_ratio_)])
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.plot(np.array(dimensions['norm']).transpose()[:,0,:]); plt.title('Bins normalizados')
plt.subplot(1,2,2)
plt.plot(np.array(dimensions['crop']).transpose()[:,0,:]); plt.title('Bins cortados')
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.plot(np.array(dimensionsLate['norm']).transpose()[:,0,:]); plt.title('Bins normalizados')
plt.subplot(1,2,2)
plt.plot(np.array(dimensionsLate['crop']).transpose()[:,0,:]); plt.title('Bins cortados')
```
