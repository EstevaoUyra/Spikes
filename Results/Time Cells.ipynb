```python
from spikeHelper.visuals import heatAct
from spikeHelper.dataOrganization import trialToXyT, XyTfromEpoch
import pickle
%matplotlib inline
import matplotlib.pyplot as plt
import pickle
```

# Rato 7

```python
rat = 7
data = pickle.load(open('Data/r'+str(rat)+'_bin50_sigma100.pickle','rb'))
data = XyTfromEpoch(data,getBins=[10,36])[:,:20,:]
print('Number of neurons',data.shape[0])
print('Number of times',data.shape[1])
print('Number of trials',data.shape[2])
```

```python
plt.subplot(1,2,1)
heatAct(trialToXyT(data[:,:,:100]))
plt.subplot(1,2,2)
heatAct(trialToXyT(data[:,:,-100:]))
```
