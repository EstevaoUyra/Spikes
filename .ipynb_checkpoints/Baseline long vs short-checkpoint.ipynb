# Predict type of trial from baseline 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from spikeHelper.dataOrganization import *
from sklearn.svm import SVC
from sklearn.metrics import f1_score
import optunity
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import make_scorer, confusion_matrix
from spikeHelper.loadSpike import rat as ratWrapper

%matplotlib inline
```

```python
scores = {}
surr = {}
for rat in [7,8,9,10]:
    short = ratWrapper(rat, 100, 50)
    short.selecTrials({'maxDuration': 700})
    short.prepareXyT(-500,0)
    
    long = ratWrapper(rat, 100, 50)
    long.selecTrials({'minDuration': 1200})
    long.prepareXyT(tmin = -500, tmax = 0)
    
    frShort = short.X.mean(axis=1)
    frLong = long.X.mean(axis=1)
    
    X = np.vstack((frShort.transpose(),frLong.transpose()))
    y = np.hstack((np.zeros(frShort.shape[1]),np.ones(frLong.shape[1])))
    
    cv_decorator = optunity.cross_validated(x=X, y=y, num_folds=5)
    def svm_rbf_tuned_auroc(x_train, y_train, x_test, y_test, C, logGamma):
        model = SVC(C=C, gamma=10 ** logGamma,class_weight= 'balanced').fit(x_train, y_train)
        y_pred = model.predict(x_test)
        auc = f1_score(y_test, y_pred,average='macro')
    optimal_rbf_pars, info, _ = optunity.maximize(svm_rbf_tuned_auroc, num_evals=150, C=[3, 10], logGamma=[-6, -2.5])
    print("Optimal parameters: " + str(optimal_rbf_pars))

    scores[rat] = []
    surr[rat] = []
    longMultiplicity = 2
    for i in range(frLong.shape[1]-frShort.shape[1]*longMultiplicity):
        
        X = np.vstack((frShort.transpose(),frLong[:,i:i+frShort.shape[1]*longMultiplicity].transpose()))
        y = np.hstack((np.zeros(frShort.shape[1]),np.ones(frShort.shape[1]*longMultiplicity )))
        Xtrain,Xtest,ytrain,ytest = train_test_split(X,y)
        
        clf = SVC(class_weight= 'balanced',C = optimal_rbf_pars['C'],gamma = 10**optimal_rbf_pars['logGamma'])
        clf.fit(Xtrain,ytrain)
        
        ypred = clf.predict(Xtest)
        scores[rat].append(f1_score(ytest,ypred,average='macro'))
        surr[rat].append(f1_score(ytest,np.random.rand(ytest.shape[0]) < longMultiplicity/(longMultiplicity+1),average= 'macro' ) )
```

```python
plt.figure(figsize=(10,10))
for rat in [7,8,9,10]:
    plt.subplot(4,1,rat-6)
    plt.plot(scores[rat])
    plt.axhline(np.array(surr[rat]).mean(),color='r')
    plt.axhline(np.array(surr[rat]).mean()-np.array(surr[rat]).std(),color='r',linestyle='--')
    plt.axhline(np.array(surr[rat]).mean()+np.array(surr[rat]).std(),color='r',linestyle='--')
```
